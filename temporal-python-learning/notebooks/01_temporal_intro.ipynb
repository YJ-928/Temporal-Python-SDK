## What is Temporal?

Temporal is an open-source **durable execution platform**. It guarantees that your application code runs to completion, even when failures occur (network outages, server crashes, bugs).

### Key Idea: Durable Execution
- Every step of your Workflow is persisted as an **event** in the event history
- If a failure occurs, Temporal **replays** the event history to restore state
- Activities are **not re-executed** during replay — cached results are used

## Core Concepts

| Component | Description |
|-----------|-------------|
| **Workflow** | Durable function that orchestrates tasks |
| **Activity** | Non-deterministic work (API calls, DB writes) |
| **Worker** | Process that executes Workflows and Activities |
| **Task Queue** | Named queue that routes work to Workers |
| **Temporal Cluster** | Server that manages orchestration and history |

## Architecture Diagram

```
┌─────────────────┐       ┌─────────────────────┐       ┌──────────────┐
│   Your Code     │◄─────►│  Temporal Cluster    │◄─────►│   Database   │
│   (Workers)     │  gRPC │   (Server)           │       │ (Event       │
│                 │       │                      │       │  History)    │
└─────────────────┘       └─────────────────────┘       └──────────────┘
```

- **Workers** run YOUR code (Workflows + Activities)
- **Temporal Cluster** manages orchestration — it never runs your code
- **Database** stores the event history for durability

## Temporal CLI Commands

```bash
# Start dev server
temporal server start-dev

# List workflows
temporal workflow list

# Start a workflow
temporal workflow start --type GreetSomeone --task-queue greeting-tasks --input '"Alice"'

# Show workflow history
temporal workflow show --workflow-id <id>
```

## Summary

- Temporal provides **durable execution** — your code always runs to completion
- **Workflows** orchestrate; **Activities** do the work
- **Workers** execute your code; the **Cluster** manages orchestration
- Use `temporal server start-dev` to start developing locally
